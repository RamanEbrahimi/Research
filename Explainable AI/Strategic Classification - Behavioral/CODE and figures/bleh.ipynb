```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def p_exp(z, gamma):
    return np.exp(-(-np.log(z))**gamma)

def p_z(z):
    return z

# Define the range of z values and gamma values
z_values = np.linspace(0.01, 1, 500)  # Avoid log(0) with z > 0
gammas = [0.3, 0.5, 0.7, 0.9]


# Plot the functions for each gamma
plt.figure(figsize=(10, 6))

for gamma in gammas:
    plt.plot(z_values, p_exp(z_values, gamma), label=r'$\gamma$ = {}'.format(gamma))

# Plot p(z) = z
plt.plot(z_values, p_z(z_values), label="p(z) = z", linestyle="--", color="black")

# Add plot details
plt.xlabel("z")
plt.ylabel("p(z)")
plt.title("Comparison of p(z) = exp(-(-ln(z))^gamma) and p(z) = z")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the functions
def p_exp(z, gamma):
    return np.exp(-(-np.log(z))**gamma)

def p_z(z):
    return z

# Define the range of z values and gamma values
z_values = np.linspace(0.01, 1, 500)  # Avoid log(0) with z > 0
gammas = [0.3, 0.5, 0.7, 0.9]

# Prepare the data for Plotly
fig = go.Figure()

# Add p(z) = exp(-(-ln(z))^gamma) for multiple gammas
for gamma in gammas:
    fig.add_trace(go.Scatter(
        x=z_values,
        y=p_exp(z_values, gamma),
        mode='lines',
        name=u"\u03B3 = " f"{gamma}"
    ))

# Add p(z) = z
fig.add_trace(go.Scatter(
    x=z_values,
    y=p_z(z_values),
    mode='lines',
    name="p(z) = z",
    line=dict(dash='dash', color='black')
))

# Update layout
fig.update_layout(
    legend=dict(
        font=dict(
            color="black",  # Change legend text color
            size=12        # Adjust font size if needed
        )
    ),
    showlegend=True,
    font=dict(size=20),  # Increase the font size for the entire plot
    height=600,
    width=800,
    template='presentation',
    plot_bgcolor="rgba(0, 0, 0, 0)",  # Transparent plot background
    paper_bgcolor="rgba(0, 0, 0, 0)",  # Transparent page background
    xaxis=dict(
        title="z",
        showgrid=False,  # Remove grid for x-axis
        color = "black"  # Change the color of the x-axis labels to white
    ),
    yaxis=dict(
        title="p(z)",
        showgrid=False,  # Remove grid for y-axis
        color = "black"  # Change the color of the y-axis labels to white
    ),
)

# Show the plot
fig.show()
```
